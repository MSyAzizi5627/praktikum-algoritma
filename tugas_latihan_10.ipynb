Tuples

```python
#a.Buat sebuah tuple bernama mahasiswa yang berisi data diri anda
# dengan 3 nilai: (“Nama”, “NIM”,”Prodi”)

mahasiswa = ("Muhammad Syafiq Azizi", "2505060045", "Teknologi informasi")
print("Tuple awal:", mahasiswa)

#b. ubah prodi menjadi teknologi informasi dan elektro
mahasiswa_list = list(mahasiswa)
mahasiswa_list[2] = "Teknologi Informasi dan Elektro"
mahasiswa = tuple(mahasiswa_list)
print("Tuple setelah ubah prodi:", mahasiswa)

#Tampilkan data NIM dan Prodi
nim = mahasiswa[1]
prodi = mahasiswa[2]
print("\nData yang diminta:")
print("NIM  :", nim)
print("Prodi:", prodi)

```

Dictionary

```python
#a.Buat dictionary keranjang
keranjang ={
    "apel": 5000,
    "pisang": 3000,
    "jeruk": 7000,
    "anggur": 15000
}

#b.&c. Gunakan perulangan untuk menampilkan setiap buah & harga dan
#      menghitung total
total_harga = 0
print("\nDaftar harga buah:")
for buah, harga in keranjang.items():
    print(f"Harga {buah}: Rp {harga}")
    total_harga += harga

#d.Tampilkan harga total belanja
print(f"total belanja adalah Rp{total_harga}")

```
