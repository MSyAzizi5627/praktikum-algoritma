```python
import math

# Luas Persegi Panjang
def luas_persegi_panjang(p, l):
  hasil = p * l
  return hasil

p = luas_persegi_panjang(5,4)
print(f"luas persegi panjang adalah = {p}")

# Volume persegi Panjang
def volume_persegi_panjang(p,l,t):
  hasil = p * l * t
  return hasil

t = volume_persegi_panjang(8,5,4)
print(f"volume persegi panjang adalah = {t}")

# Luas Lingkaran
def luas_lingkaran(r):
  hasil = math.pi*r**2
  return hasil

l = luas_lingkaran(10)
print(f"luas lingkaran adalah = {l}")

# Volume Bola
def volume_bola(r):
  hasil = (4/3)*math.pi*r**3
  return hasil

b = volume_bola(7)
print(f"volume bola adalah = {b}")
```

```python
import math

# Luas Persegi Panjang
def luas_persegi_panjang(p, l):
  return p * l

# Volume balok
def volume_balok(p,l,t):
  return p * l * t

# Luas Lingkaran
def luas_lingkaran(r):
  return math.pi*r**2

# Volume Bola
def volume_bola(r):
  return (4/3)*math.pi*r**3

# Menu utama
print("=== OPERASI PERHITUNGAN MATEMATIKA ===")
print("1. Luas Persegi Panjang")
print("2. Volume balok")
print("3. Luas Lingkaran")
print("4. Volume Bola")

pilihan = input("Pilih operasi (1/2/3/4): ")

if pilihan == "1":
    p = int(input("Masukkan panjang: "))
    l = int(input("Masukkan lebar: "))
    print(f"Luas persegi panjang = {luas_persegi_panjang(p, l)}")

elif pilihan == "2":
    p = int(input("Masukkan panjang: "))
    l = int(input("Masukkan lebar: "))
    t = int(input("Masukkan tinggi: "))
    print(f"Volume balok = {volume_balok(p, l, t)}")

elif pilihan == "3":
    r = float(input("Masukkan jari-jari lingkaran: "))
    print(f"Luas lingkaran = {luas_lingkaran(r)}")

elif pilihan == "4":
    r = float(input("Masukkan jari-jari bola: "))
    print(f"Volume bola = {volume_bola(r)}")

else:
    print("Pilihan tidak ada!")
```

```python
def luas_persegi_panjang():
    print("Menghitung luas persegi panjang")

    panjang = int(input("Panjang\t:"))
    lebar = int(input("Lebar\t:"))

    return panjang * lebar

def volume_balok():
    print("Menghitung volume balok")

    panjang = int(input("Panjang\t:"))
    lebar = int(input("Lebar\t:"))
    tinggi = int(input("Tinggi\t:"))
    return panjang * lebar * tinggi

def luas_lingkaran():
    print("Menghitung luas lingkaran")

    jari_jari = int(input("Jari-jari\t:"))
    return 3.14 * jari_jari**2

def volume_bola():
    print("Menghitung volume bola")

    jari_jari = int(input("Jari-jari\t:"))
    return .75 * 3.14 * jari_jari**3

# Contoh pemanggilan fungsi
print("Luas Persegi Panjang:", luas_persegi_panjang())
print("Volume Balok:", volume_balok())
print("Luas Lingkaran:", luas_lingkaran())
print("Volume Bola:", volume_bola())
```
